```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
headers = {'User-Agent':'GoogleBot'}
```

```python
# get: 請求，傳給伺服端，並下載
# 正規表達式，數字:[0-9] 、 {n, m}: 出現n~m次
# set: 集合，不會包含重複的資料
# .text會取得html文字
# re.findall(a,b) : 在b裡面尋找全部的a
urls = []
for i in range(1000):
    page = "https://www.104.com.tw/company/search/?area=6001014000&jobsource=checkc&mode=s&page={}".format(str(i+1)) #台南市 
    resp = requests.get(page, headers = headers)
    nurls = list(set(re.findall('//www.104.com.tw/company/\w{1,}', resp.text))) 
    if len(urls) == len(list(set(urls + nurls))):
        break
    urls = list(set(urls + nurls))
    print(len(urls))
```

```python
def BasicInfo(soup):
    #公司資訊
    industry_category, contact, industry_describe, telephone, capital, fax, employee, adress, website= ['']*9
    information = []

    for i in soup.find_all('div',{'class':'row mb-2'}):
        information.extend(i.text.split(' '))

    while '' in information:
        information.remove('')

    if '經濟部商業司查詢' in information:
        information.remove('經濟部商業司查詢')

    # print(information)

    if '產業類別' in information:
        index = information.index('產業類別')
        industry_category = information[index+1]

    if '聯絡人' in information:
        index = information.index('聯絡人')
        contact = information[index+1]

    if '產業描述' in information:
        index = information.index('產業描述')
        industry_describe = information[index+1]

    if '電話' in information:
        index = information.index('電話')
        telephone = information[index+1]

    if '資本額' in information:
        index = information.index('資本額')
        capital = information[index+1]

    if '傳真' in information:
        index = information.index('傳真')
        fax = information[index+1]

    if '員工人數' in information:
        index = information.index('員工人數')
        employee = information[index+1]

    if '地址' in information:
        index = information.index('地址')
        adress = information[index+1]

    if '公司網址' in information:
        index = information.index('公司網址')
        website = information[index+1]
        
    return industry_category, contact, industry_describe, telephone, capital, fax, employee, adress, website
```

```python
def url2df(url):
    print('https:' + url + '?jobsource=checkc')
    resp = requests.get('https:' + url + '?jobsource=checkc')
    soup = BeautifulSoup(resp.text) #傳入取回的HTML結構字串
    #公司名稱
    comp_name = soup.select('h1.d-inline')[0].text
    #公司資訊
    industry_category, contact, industry_describe, telephone, capital, fax, employee, adress, website = BasicInfo(soup)

    ndf = pd.DataFrame([{'公司名稱':comp_name,
                    '產業類別': industry_category,
                    '聯絡人': contact,
                    '產業描述': industry_describe,
                    '電話': telephone,
                    '資本額': capital,
                    '傳真': fax,
                    '員工人數': employee,
                    '地址': adress,
                    '公司網址': website}],
                columns=['公司名稱', '產業類別', '聯絡人', '產業描述', '電話', '資本額', '傳真', '員工人數', '地址', '公司網址'])
    # print(ndf)
    
    return ndf
```

```python
df = []
fail = []
ndf = pd.DataFrame()
for i, url in enumerate(urls): #组合為一個索引序列，i是索引、url是值(網址)
    if i % 100 == 0:
        print(i)
    try:
        ndf = url2df(url)
        df.append(ndf)
    except:
        fail.append(url)
df = pd.concat(df, ignore_index=True)
df
```
